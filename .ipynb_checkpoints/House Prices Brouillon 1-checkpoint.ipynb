# Regression Linéaire SkLearn

```python
SalePrice_1 = df.loc[:,'SalePrice']
SalePrice = SalePrice_1.values
print(SalePrice)
```

```python
FirstFloor_1 = df.loc[:,'1stFlrSF']
FirstFloor = FirstFloor_1.values
print(FirstFloor)
```

```python
X = np.array([SalePrice]).reshape((-1, 1))
y = np.array([FirstFloor]).reshape((-1, 1))
```

```python
reg = LinearRegression().fit(X, y)
reg.score(X, y)
```

```python
reg.coef_[0][0]
```

```python
reg.intercept_[0]
```

```python
reg.coef_[0][0] * 208000 + reg.intercept_[0]
```
